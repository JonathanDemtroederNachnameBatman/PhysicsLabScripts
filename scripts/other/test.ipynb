```python
import numpy as np
import matplotlib.pyplot as plt
```

# Logistic Map
$x_{k+1}=ax_k(1-x_k)$

```python
x = np.ones(100)
a = 3.5
x[0] = 0.12345

for k in range(len(x)-1):
    x[k+1] = a * x[k] * (1 - x[k])

plt.plot(x, '.-')
plt.grid()
plt.show()
```

# Henon Map
$x_{k+1}=1-ax_k^2+y_k$

$y_{k+1}=bx_k$

```python
n = 1000000
x = np.ones(n)
y = np.ones(n)
a = 1.4
b = 0.3
x[0] = 0.5
y[0] = 0.5

for k in range(len(x)-1):
    x[k+1] = 1 - a * x[k]**2 + y[k]
    y[k+1] = b * x[k]

plt.plot(x, y, '.', markersize=0.01)
plt.grid()
plt.show()
```
