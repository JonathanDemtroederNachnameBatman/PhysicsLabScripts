# Ordinary Differential Equation

```python
import numpy as np
import matplotlib.pyplot as plt

n = 200
h = 0.1
tt = np.arange(n)*h
ex = np.exp(tt * 1j)
```

### Euler method
Harmonic Oszillator
$\frac{dx(t)}{dt}=ix(t)$

```python
x = np.full((n), 0 + 0j)
x[0] = 1

for k in range(n-1):
    x[k+1] = x[k] + 1j*h*x[k]

plt.plot(tt, np.real(x), '.', label='Real')
plt.plot(tt, np.imag(x), '.', label='Imag')
plt.plot(tt, np.abs(x), '.', label='Abs')
plt.plot(tt, ex, '.', label='Analytic Solution')
plt.legend()
plt.grid()
```

### Leap-Frog method

```python
x = np.full((n), 0 + 0j)
x[0] = 1
x[1] = 1 + 1j*h

for k in np.arange(1,n-1):
    x[k+1] = x[k-1] + 1j*2.*h*x[k]

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.plot(tt, np.real(x), '.', label='Real')
ax1.plot(tt, np.imag(x), '.', label='Imag')
ax2.plot(tt, np.abs(x), '-', label='Abs')
ax1.plot(tt, ex, '.', label='Analytic Solution')
#plt.legend()
ax2.grid()
ax1.grid()
```
