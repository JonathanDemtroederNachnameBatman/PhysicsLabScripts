# Mandelbrot
$z_{n+1}=z_n^2+c$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 1001)

#z = np.meshgrid(x, y)
#print(z)
```

```python

n_arr = np.zeros((len(x), len(x)))

iterations = 100

for i in range(len(x)):
    for j in range(len(x)):
        c = x[i] + x[j] * 1j
        z = np.zeros(iterations, dtype='complex128')
        for k in range(1, len(z)):
            z[k] = z[k-1]**2 + c

            if np.abs(z[k]) >= 2:
                n_arr[i][j] = k * 2.1
                break

            n_arr[i][j] = 2 * iterations

```

```python
fig, ax = plt.subplots()

ax.contourf(x, x, n_arr.T)
```
